## Description

Import the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

Load the data set

```python
df = pd.read_csv('Crypto_Data.csv')
```

Show the data

```python
df
```

Set the date as index

```python
df = df.set_index(pd.DatetimeIndex(df['Date'].values))
```

Drop (remove) date column

```python
df = df.drop('Date', axis=1)
```

```python
df
```

Get some statistics on the data

```python
df.describe()
```

Visualize and graph the cryptocurrency price

```python
my_cripto = df
plt.figure(figsize=(12.2, 4.5))
for c in my_cripto.columns.values:
    plt.plot(my_cripto[c], label=c)
    
plt.title('Cryptocurrency Graph')
plt.xlabel('Years')
plt.ylabel('Crypto Price USD ($)')
plt.legend(my_cripto.columns.values, loc='upper left')
plt.show()
```

Scale the data

```python
from sklearn import preprocessing
```

```python
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,100))
```

```python
scaled = min_max_scaler.fit_transform(df)
```

Show the data

```python
scaled
```

Convert the scaled data into a data frame

```python
df_scale = pd.DataFrame(scaled, columns=df.columns)
```

Set the date as index

```python
df_scale = df_scale.set_index(pd.DatetimeIndex(df.index))
```

Show the data

```python
df_scale
```

Visualize and graph the cryptocurrency price

```python
my_cripto = df_scale
plt.figure(figsize=(12.2, 4.5))
for c in my_cripto.columns.values:
    plt.plot(my_cripto[c], label=c)
    
plt.title('Cryptocurrency Scaled Graph')
plt.xlabel('Years')
plt.ylabel('Individual Cryptos Scaled Price from Min To Max')
plt.legend(my_cripto.columns.values, loc='upper left')
plt.show()
```

Get the daily simple return

```python
DSR = df.pct_change(1)
```

Show the data

```python
DSR
```

Visualize and graph the simple daily return

```python
my_cripto = DSR
plt.figure(figsize=(12.2, 4.5))
for c in my_cripto.columns.values:
    plt.plot(my_cripto[c], label=c, alpha=.7, lw=2)
    
plt.title('Daily Simple Return')
plt.xlabel('Years')
plt.ylabel('Percentage (in decimal form)')
plt.legend(my_cripto.columns.values, loc='upper left')
plt.show()
```

Print the cryptocurrencies volatility

```python
print('The cryptocurrency volatility:')
DSR.std()
```

Show mean (average) daily simple return

```python
DSR.mean()
```

Get the correlation matrix

```python
DSR.corr()
```

```python
import seaborn as sns
```

```python
plt.subplots(figsize=(11,11))
sns.heatmap(DSR.corr(), annot=True, fmt=".2%")
```

Get and show the daily cumulative simple returns

```python
DSCR = (DSR+1).cumprod()
```

Show the data

```python
DSCR
```

Visualize and graph the cumulative daily return

```python
my_cripto = DSCR
plt.figure(figsize=(12.2, 4.5))
for c in my_cripto.columns.values:
    plt.plot(my_cripto[c], label=c)
    
plt.title('Daily Cumulative Simple Return')
plt.xlabel('Years')
plt.ylabel('Growth of $1 Investment')
plt.legend(my_cripto.columns.values, loc='upper left')
plt.show()
```

This program determines the worst and best months for return using Python

Import the libraries

```python
import pandas as pd
import numpy as np
```

Load the data

```python
df = pd.read_csv("SP500.csv")
```

Show the data

```python
df
```

Set the date as the index

```python
df = df.set_index(pd.DatetimeIndex(df['Date']))
```

Show the data

```python
df
```

Create lists to store the returns of each day in that month

```python
Jan = []
Feb = []
Mar = []
Apr = []
May = []
June = []
July = []
Aug = []
Sept = []
Oct = []
Nov = []
Dec = []
```

Create a function to get the returns for each month and append it to the corresponding list

```python
def get_returns(month, returns):
    if month == '01':
        Jan.append(float(returns))
    elif month == '02':
        Feb.append(float(returns))
    elif month == '03':
        Mar.append(float(returns))
    elif month == '04':
        Apr.append(float(returns))
    elif month == '05':
        May.append(float(returns))
    elif month == '06':
        June.append(float(returns))
    elif month == '07':
        July.append(float(returns))
    elif month == '08':
        Aug.append(float(returns))
    elif month == '09':
        Sept.append(float(returns))
    elif month == '10':
        Oct.append(float(returns))
    elif month == '11':
        Nov.append(float(returns))
    elif month == '12':
        Dec.append(float(returns))
    else:
        print('Something is wrong with the input for the month')
    
```

Calculate show the daily simple returns

```python
DSR = df['Close'].pct_change(1)
```

```python
df['DSR'] = DSR
```

Show the data

```python
df
```

Remove first row

```python
df = df[1:]
```

Show the data

```python
df
```

Create a loop to gather the daily simple returns of each month from the data set and append them to the lists

```python
for i in range(0, len(df)):
    df_date = str(df.index[i])
    df_returns = df['DSR'][i]
    df_month = df_date.split('-')[1]
    # Add the returns to the list corresponding to that month
    get_returns(df_month, df_returns)
```

Create a function to average the returns of each month

```python
def AVG(month):
    return [sum(month) / len(month)]
```

Create a new dataframe

```python
df_AVG = pd.DataFrame()
```

Get the average returns for each month and add the values under a new column 'AVG'

```python
df_AVG['AVG'] = AVG(Jan) + AVG(Feb) + AVG(Mar) + AVG(Apr) + AVG(May) + AVG(June) + AVG(July) + AVG(Aug) + AVG(Sept) + AVG(Oct) + AVG(Nov) + AVG(Dec)
```

Set index to be the corresponding integer value of the month 

```python
df_AVG = df_AVG.set_index(df_AVG.index + 1)
```

Show the average monthly returns

```python
df_AVG * 100
```

Plot the Average Monthly Returns

```python
df_AVG.plot.bar()
```

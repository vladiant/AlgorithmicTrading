```python
start_date = "2021-05-08"
end_date="2021-05-09"
topic = "cryptocurrency"
NEWSAPI_TOKEN = "YOUR_API_KEY"
```

```python
api_url = f'https://newsapi.org/v2/everything?q={topic}&from={start_date}&to={end_date}&sortBy=publishedAt&apiKey={NEWSAPI_TOKEN}'
```

```python
data = requests.get(api_url)
```

```python
print(data.status_code)
```

```python
data.json()
```

```python
data.json()['articles'][0]['publishedAt']
```

```python
import pandas as pd
```

```python
df = pd.read_json(api_url)
```

```python
df.articles
```

```python
df_articles = pd.json_normalize(data.json(), record_path=['articles'])
```

```python
df_articles
```

```python
df_articles.to_csv('CryptoCurrencyNews.csv')
```

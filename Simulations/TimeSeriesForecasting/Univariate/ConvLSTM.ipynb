Import the libraries

```python
import pandas as pd
import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Flatten, ConvLSTM2D

import matplotlib.pyplot as plt
```

Split an univariate sequence into samples

```python
def split_sequence(sequence, n_steps):
    X, y = list(), list()
    for i in range(len(sequence)):
        # Find the end of the pattern
        end_ix = i + n_steps
        # Check if we are bound by sequence
        if end_ix > len(sequence) - 1:
            break
        # Gather input and output parts of the pattern
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    
    return np.array(X), np.array(y)
```

Load the data

```python
df = pd.read_csv("SP500.csv")
```

Show the data

```python
df
```

Add indices for every day in the period

```python
start_date = df.Date.min()
end_date = df.Date.max()
```

```python
df_date = pd.date_range(start=start_date, end=end_date).to_frame(name="Date")
df_date.reset_index(drop=True, inplace=True) 
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df = df.merge(df_date, how='right')
```

Interpolate missing values

```python
df = df.interpolate(method='linear', limit_direction='both')
df
```

Define input sequence

```python
raw_seq = np.array(df.Close)
```

Choose a number of time steps

```python
n_total_steps = 32
n_features = 1
```

```python
n_steps = 8
n_seq = n_total_steps // n_steps
```

Split into samples

```python
X, y = split_sequence(raw_seq[:-n_total_steps*n_seq], n_total_steps)
```

Reshape from [samples, timesteps] into [samples, timesteps, rows, columns, features]

```python
X = X.reshape((X.shape[0], n_seq, 1, n_steps, n_features))
```

Set test data

```python
X_test, y_test = split_sequence(raw_seq[-n_total_steps*n_seq:], n_total_steps) 
X_test = X_test.reshape((X_test.shape[0], n_seq, 1, n_steps, n_features))
```

Define model

```python
model = Sequential()
model.add(ConvLSTM2D(filters=64, kernel_size=(1,2), activation='relu', input_shape=(n_seq, 1, n_steps, n_features)))
model.add(Flatten())
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mse')
```

Fit model

```python
model.fit(X, y, epochs=200)
```

Demonstrate training

```python
yhat = model.predict(X)
```

```python
plt.plot(y)
plt.plot(yhat)
plt.show()
```

Demonstrate prediction

```python
yhat = model.predict(X_test)
```

```python
plt.plot(y_test, label="Expected")
plt.plot(yhat, label="Predicted")
plt.legend()
plt.show()
```

```python
print(f"Predicted: {yhat[0][0]:.2f}  Expected: {y_test[0]:.2f}")
```
